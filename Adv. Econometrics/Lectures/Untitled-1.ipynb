# Julia First Touch

```julia
# requirements
using LinearAlgebra
using Statistics
using Plots
using Distributions
```

## Random Variables

```julia
n = 30
xbar = Vector{Float64}(undef, 200000)
for j in 200000
    x = randn(n)
    xbar[j] = mean(x)
end
```

## Computing Statistics

```julia
mean(xbar)
var(xbar)
sqrt(n).*xbar
```

```julia

histogram(sqrt(n).*(xbar), bins=100, normed=true)
```
